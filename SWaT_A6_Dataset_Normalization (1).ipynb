```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

file_path = '/content/drive/MyDrive/Dec2019.xlsx'

df = pd.read_excel(file_path)

minmax_scaled_data = df.select_dtypes(include=['float64', 'int64']).columns

if not minmax_scaled_data.empty:
    # Handle missing values
    df[minmax_scaled_data] = df[minmax_scaled_data].fillna(0)
    # Apply MinMaxScaler
    scaler = MinMaxScaler()
    df[minmax_scaled_data] = scaler.fit_transform(df[minmax_scaled_data])
else:
    print("No numeric columns found for scaling.")

# Style and display the table
df_head = df.head(25).style.set_table_styles(
    [{'selector': 'thead th',
      'props': [('font-weight', 'bold'), ('background-color', '#4CAF50'), ('color', 'white')]},
     {'selector': 'tbody td',
      'props': [('border', '1px solid black'), ('padding', '8px')]},
     {'selector': 'table',
      'props': [('border-collapse', 'collapse'), ('width', '100%')]},
     {'selector': 'thead th',
      'props': [('text-align', 'center')]}])
df_head
```
